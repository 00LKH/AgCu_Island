```python
from ase.visualize import view
from ase.io import read

# 1. DB 로드 후 저장된 구조 리스트 가져오기 (ASE DB 표준 로드)
candidates = read("../03_gcga/process/gcga.db", index=":")

# 데이터가 잘 로드되었는지 간단히 확인
print(f"로드된 구조 개수: {len(candidates)}개")
print(f"최저 에너지: {min([a.get_potential_energy() for a in candidates]):.3f} eV")

# 2. nglview 시각화
# ASE의 view_ngl이 NGLDisplay.gui (HBox)를 반환하며, control_box에 있는
# "Ball size" FloatSlider(기본 value=0.5)가 spacefill의 radiusScale을 제어한다.
# 슬라이더 value를 바꾸면 등록된 observer가 update_spacefill(radiusScale=..)을 자동 호출.
gui = view(candidates, viewer='ngl')
ball_size_slider = next(
    w for w in gui.control_box.children
    if getattr(w, 'description', '') == 'Ball size'
)
ball_size_slider.value = 1.0  # 기본 0.5 → 1.0
gui
```
