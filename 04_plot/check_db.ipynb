```python
from ase.visualize import view
from ase.io import read

# 1. DB 로드 후 저장된 구조 리스트 가져오기 (ASE DB 표준 로드)
candidates = read("../03_gcga/process/gcga.db", index=":")

# 데이터가 잘 로드되었는지 간단히 확인
print(f"로드된 구조 개수: {len(candidates)}개")
print(f"최저 에너지: {min([a.get_potential_energy() for a in candidates]):.3f} eV")

# 2. nglview 시각화
view(candidates, viewer='ngl')
```
